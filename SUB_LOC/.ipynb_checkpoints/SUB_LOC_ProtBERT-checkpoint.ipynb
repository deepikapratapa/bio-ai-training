# 🧬 Dr. Aris and the Geography of Proteins

Dr. Aris had already gathered the data with great care.

From the vast archive of protein knowledge, he had selected curated sequences, cleaned them, simplified their cellular destinations, and split them into training, validation, and test sets. The map was ready.

Now came the true question:

**Could a protein language model read sequence alone and infer where that protein belongs inside the cell?**

A protein that settles into the membrane often carries hydrophobic signals.
A protein sent to the nucleus may bear localization motifs.
A protein secreted outside the cell may reveal its destiny in the very grammar of its amino acids.

Dr. Aris now turns to a transformer encoder — a model trained not on labels, but on the language of proteins itself.

He will ask it to convert each sequence into an embedding:
a compact numerical signature of biological meaning.

Then, with a downstream classifier, he will test whether those embeddings contain enough information to predict subcellular localization.

## Mission of this notebook

In this notebook, Dr. Aris will:

1. Load the prepared train, validation, and test sets
2. Encode localization labels into integers
3. Load a protein language model
4. Generate sequence embeddings
5. Train a downstream classifier
6. Evaluate localization prediction performance
7. Visualize the confusion matrix
8. Inspect whether the embedding space captures cellular geography

## Imports

```python
import os
import re
import time
import random
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)

import torch
from transformers import AutoTokenizer, AutoModel
from tqdm.auto import tqdm
import umap
```

## Configuration

```python
# =========================
# Configuration
# =========================
RANDOM_STATE = 42

DATA_DIR = "./subcellular_localization_data_prep"
TRAIN_PATH = f"{DATA_DIR}/train.csv"
VAL_PATH   = f"{DATA_DIR}/val.csv"
TEST_PATH  = f"{DATA_DIR}/test.csv"

MODEL_NAME = "Rostlab/prot_bert"   # later: ProtT5 / ESM-2
MAX_LEN = 1024
BATCH_SIZE = 8

CLASSIFIER_NAME = "logreg"   # options: "logreg", "mlp"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)
torch.manual_seed(RANDOM_STATE)

print("Device:", DEVICE)
print("Model:", MODEL_NAME)
print("Classifier:", CLASSIFIER_NAME)
```

## Opening the prepared datasets

Dr. Aris begins by reopening the three carefully prepared splits:
- the training set, where the model will learn,
- the validation set, where choices can be checked,
- and the test set, where the final judgment will be made.

Because these splits were created earlier in a stratified way, they preserve the balance of the localization classes.

```python
train_df = pd.read_csv(TRAIN_PATH)
val_df   = pd.read_csv(VAL_PATH)
test_df  = pd.read_csv(TEST_PATH)

print("Train shape:", train_df.shape)
print("Val shape:  ", val_df.shape)
print("Test shape: ", test_df.shape)

train_df.head()
```

## Inspect columns

```python
print("Train columns:", train_df.columns.tolist())
```

## Check class distribution

```python
print("Train distribution:")
print(train_df["localization"].value_counts())

print("\nValidation distribution:")
print(val_df["localization"].value_counts())

print("\nTest distribution:")
print(test_df["localization"].value_counts())
```

## Translating destinations into machine-readable targets

The proteins still carry labels in biological language:
*Cytoplasm*, *Nucleus*, *Membrane*, and so on.

But the classifier needs numerical targets.

So Dr. Aris now builds a label encoder, preserving the mapping carefully so that every future prediction can still be translated back into cell biology.

## Label encoding

```python
label_encoder = LabelEncoder()

label_encoder.fit(train_df["localization"])

train_df["label"] = label_encoder.transform(train_df["localization"])
val_df["label"]   = label_encoder.transform(val_df["localization"])
test_df["label"]  = label_encoder.transform(test_df["localization"])

label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
id_to_label = {v: k for k, v in label_mapping.items()}

print("Label mapping:")
print(label_mapping)
```

## Prepare Sequences

```python
def clean_sequence(seq):
    seq = str(seq).upper().strip()
    seq = re.sub(r"[^ACDEFGHIKLMNPQRSTVWYUZOBX]", "", seq)
    seq = re.sub(r"[UZOB]", "X", seq)
    return seq

def format_for_protbert(seq):
    return " ".join(list(seq))

train_df["sequence_clean"] = train_df["sequence"].apply(clean_sequence)
val_df["sequence_clean"]   = val_df["sequence"].apply(clean_sequence)
test_df["sequence_clean"]  = test_df["sequence"].apply(clean_sequence)

train_df["sequence_fmt"] = train_df["sequence_clean"].apply(format_for_protbert)
val_df["sequence_fmt"]   = val_df["sequence_clean"].apply(format_for_protbert)
test_df["sequence_fmt"]  = test_df["sequence_clean"].apply(format_for_protbert)

train_df[["sequence", "sequence_fmt"]].head(2)
```

## Asking the transformer to read the proteins

Now comes the heart of the experiment.

Dr. Aris loads a protein language model — not yet to fine-tune it, but to use it as a reader.

Each protein sequence will pass through the transformer.
From this, the model will produce token-level representations.
These token representations will then be pooled into a single vector:
an embedding that summarizes the sequence in numerical form.

If these embeddings truly capture biologically meaningful structure, then cellular localization may already be encoded within them.

## Load tokenizer and model

```python
print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, do_lower_case=False)

print("Loading model...")
encoder_model = AutoModel.from_pretrained(
    MODEL_NAME,
    low_cpu_mem_usage=True
).to(DEVICE)

encoder_model.eval()
print("Loaded successfully.")
```

## Embedding helpers

```python
def mean_pool(last_hidden_state, attention_mask):
    mask = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
    masked_embeddings = last_hidden_state * mask
    summed = masked_embeddings.sum(dim=1)
    counts = mask.sum(dim=1).clamp(min=1e-9)
    return summed / counts

def embed_sequences(sequences, model, tokenizer, device="cpu", batch_size=8, max_len=1024):
    all_embeddings = []

    for i in tqdm(range(0, len(sequences), batch_size)):
        batch = sequences[i:i+batch_size]

        encoded = tokenizer(
            batch,
            padding=True,
            truncation=True,
            max_length=max_len,
            return_tensors="pt"
        )

        encoded = {k: v.to(device) for k, v in encoded.items()}

        with torch.no_grad():
            outputs = model(**encoded)
            pooled = mean_pool(outputs.last_hidden_state, encoded["attention_mask"])

        all_embeddings.append(pooled.cpu().numpy())

    return np.vstack(all_embeddings)
```

## Generate Embeddings

```python
start_time = time.time()

X_train = embed_sequences(
    train_df["sequence_fmt"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

X_val = embed_sequences(
    val_df["sequence_fmt"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

X_test = embed_sequences(
    test_df["sequence_fmt"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

embedding_time = time.time() - start_time

print("Embedding shapes:")
print("X_train:", X_train.shape)
print("X_val:  ", X_val.shape)
print("X_test: ", X_test.shape)

print(f"\nTotal embedding time: {embedding_time/60:.2f} minutes")
```

## Prepare Targets

```python
y_train = train_df["label"].values
y_val   = val_df["label"].values
y_test  = test_df["label"].values

print("y_train shape:", y_train.shape)
print("Number of classes:", len(label_encoder.classes_))
```

## Teaching a lightweight classifier

The transformer has now spoken.

Each protein has been converted into an embedding — a numerical signature shaped by the statistical language of evolution.

But embeddings alone do not make predictions.
Dr. Aris must now place a classifier on top of them.

He begins with a simple but powerful baseline:
**Logistic Regression**.

If a lightweight model can separate proteins by localization using these embeddings, that itself is evidence that the transformer has captured meaningful cellular information.

```python
if CLASSIFIER_NAME == "logreg":
    clf = LogisticRegression(
        max_iter=3000,
        random_state=RANDOM_STATE,
        class_weight="balanced"
    )

elif CLASSIFIER_NAME == "mlp":
    clf = MLPClassifier(
        hidden_layer_sizes=(512, 256),
        activation="relu",
        solver="adam",
        alpha=1e-4,
        batch_size=64,
        learning_rate_init=1e-3,
        max_iter=100,
        random_state=RANDOM_STATE
    )

else:
    raise ValueError("Unsupported CLASSIFIER_NAME. Use 'logreg' or 'mlp'.")

print(clf)
```

## Train Classifier

```python
train_start = time.time()
clf.fit(X_train, y_train)
train_time = time.time() - train_start

print(f"Training time: {train_time:.2f} seconds")
```

## Validation predictions

```python
val_preds = clf.predict(X_val)

val_acc = accuracy_score(y_val, val_preds)
val_macro_f1 = f1_score(y_val, val_preds, average="macro")
val_weighted_f1 = f1_score(y_val, val_preds, average="weighted")

print("Validation Metrics")
print("------------------")
print(f"Accuracy     : {val_acc:.4f}")
print(f"Macro F1     : {val_macro_f1:.4f}")
print(f"Weighted F1  : {val_weighted_f1:.4f}")
```

## Test predictions

```python
test_preds = clf.predict(X_test)

test_acc = accuracy_score(y_test, test_preds)
test_macro_f1 = f1_score(y_test, test_preds, average="macro")
test_weighted_f1 = f1_score(y_test, test_preds, average="weighted")

print("Test Metrics")
print("------------")
print(f"Accuracy     : {test_acc:.4f}")
print(f"Macro F1     : {test_macro_f1:.4f}")
print(f"Weighted F1  : {test_weighted_f1:.4f}")
```

## Reading the evaluation carefully

Dr. Aris does not trust accuracy alone.

Some localization classes may be easier, more frequent, or more strongly encoded than others.  
So he also inspects:

- **Macro F1**, which treats each class equally
- **Weighted F1**, which accounts for class frequency
- and later, the **confusion matrix**, which reveals where proteins are being mistaken for one another

This way, the notebook does not merely ask whether the model performs well,
but also **how** it succeeds — and where it struggles.

## Classification report

```python
print(classification_report(
    y_test,
    test_preds,
    target_names=label_encoder.classes_
))
```

## Confusion report

```python
cm = confusion_matrix(y_test, test_preds)

fig, ax = plt.subplots(figsize=(8, 8))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=label_encoder.classes_
)
disp.plot(ax=ax, xticks_rotation=45, colorbar=False)
plt.title("Confusion Matrix - Subcellular Localization")
plt.tight_layout()
plt.show()
```

## Per-class Accuracy view

```python
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(cm_norm, aspect="auto")

ax.set_xticks(np.arange(len(label_encoder.classes_)))
ax.set_yticks(np.arange(len(label_encoder.classes_)))
ax.set_xticklabels(label_encoder.classes_, rotation=45, ha="right")
ax.set_yticklabels(label_encoder.classes_)

for i in range(cm_norm.shape[0]):
    for j in range(cm_norm.shape[1]):
        ax.text(j, i, f"{cm_norm[i, j]:.2f}", ha="center", va="center")

ax.set_xlabel("Predicted label")
ax.set_ylabel("True label")
ax.set_title("Normalized Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=RANDOM_STATE)
X_test_2d = pca.fit_transform(X_test)

umap_df = pd.DataFrame({
    "Dim1": X_test_2d[:, 0],
    "Dim2": X_test_2d[:, 1],
    "label": [id_to_label[i] for i in y_test]
})

plt.figure(figsize=(8, 6))
for label in sorted(umap_df["label"].unique()):
    subset = umap_df[umap_df["label"] == label]
    plt.scatter(subset["Dim1"], subset["Dim2"], s=15, alpha=0.7, label=label)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Test Embeddings by Subcellular Localization")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

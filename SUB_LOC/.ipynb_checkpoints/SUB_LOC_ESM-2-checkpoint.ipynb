# 🧬 Dr. Aris and the Geography of the Cell — An ESM-2 Expedition

Dr. Aris had already prepared the data with care.

He had gathered curated proteins, cleaned their sequences, simplified their cellular destinations, and divided them into training, validation, and test sets. The map of the cell now stood ready.

But a map alone does not answer a biological question.

Now he turned to **ESM-2**, a transformer trained to read the language of proteins — not through handcrafted rules, but through the statistical grammar written by evolution itself.

His question was elegant in its simplicity:

**Can a protein’s amino acid sequence reveal where it lives inside the cell?**

A membrane protein may carry hydrophobic stretches.  
A nuclear protein may bear localization signals.  
A secreted protein may carry clues of export in its opening sequence.  
And perhaps, if the transformer has learned protein language deeply enough, these signals may already be encoded in its embeddings.

So begins this ESM-2 expedition into subcellular localization prediction.

## Mission of this notebook

In this notebook, Dr. Aris will:

1. Load the prepared training, validation, and test sets
2. Encode localization labels into numeric form
3. Load the ESM-2 protein language model
4. Convert each sequence into a fixed-length embedding
5. Train a downstream classifier on those embeddings
6. Evaluate performance on validation and test sets
7. Visualize errors using a confusion matrix
8. Optionally inspect the embedding space in two dimensions

This notebook focuses on a simple but powerful idea:

**If ESM-2 embeddings capture cellular information, then even a lightweight classifier should be able to recover localization labels from them.**

## Imports

```python
import os
import re
import gc
import time
import random
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)

import torch
from transformers import AutoTokenizer, AutoModel
from tqdm.auto import tqdm
```

```python
# =========================
# Configuration
# =========================
RANDOM_STATE = 42

DATA_DIR = "./subcellular_localization_data_prep"
TRAIN_PATH = f"{DATA_DIR}/train.csv"
VAL_PATH   = f"{DATA_DIR}/val.csv"
TEST_PATH  = f"{DATA_DIR}/test.csv"

MODEL_NAME = "facebook/esm2_t30_150M_UR50D"
MAX_LEN = 1024
BATCH_SIZE = 8

CLASSIFIER_NAME = "logreg"   # options: "logreg", "mlp"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)
torch.manual_seed(RANDOM_STATE)

print("Device:", DEVICE)
print("Model:", MODEL_NAME)
print("Classifier:", CLASSIFIER_NAME)
```

## Opening the prepared datasets

Dr. Aris begins by reopening the three splits created in the preparation notebook:

- the **training set**, where the classifier will learn,
- the **validation set**, where performance will be checked,
- and the **test set**, where the final evaluation will be made fairly.

Because the data were split in a stratified manner, the localization classes remain reasonably balanced across all three sets.

```python
train_df = pd.read_csv(TRAIN_PATH)
val_df   = pd.read_csv(VAL_PATH)
test_df  = pd.read_csv(TEST_PATH)

print("Train shape:", train_df.shape)
print("Val shape:  ", val_df.shape)
print("Test shape: ", test_df.shape)

train_df.head()
```

## Inspect Columns and Class Balance

```python
print("Columns:", train_df.columns.tolist())

print("\nTrain distribution:")
print(train_df["localization"].value_counts())

print("\nValidation distribution:")
print(val_df["localization"].value_counts())

print("\nTest distribution:")
print(test_df["localization"].value_counts())
```

## Translating biology into machine-readable labels

The proteins still speak in biological language:
*Cytoplasm*, *Nucleus*, *Membrane*, *Extracellular*, and so on.

But a classifier must work with numbers.

So Dr. Aris now builds a label encoder — a careful dictionary that preserves the meaning of each class while translating it into a numeric target.

## Encode Labels

```python
label_encoder = LabelEncoder()
label_encoder.fit(train_df["localization"])

train_df["label"] = label_encoder.transform(train_df["localization"])
val_df["label"]   = label_encoder.transform(val_df["localization"])
test_df["label"]  = label_encoder.transform(test_df["localization"])

label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
id_to_label = {v: k for k, v in label_mapping.items()}

print("Label mapping:")
print(label_mapping)
```

## Sequence Cleaning

```python
def clean_sequence(seq):
    seq = str(seq).upper().strip()
    seq = re.sub(r"[^ACDEFGHIKLMNPQRSTVWYUZOBX]", "", seq)
    seq = re.sub(r"[UZOB]", "X", seq)
    return seq

train_df["sequence_clean"] = train_df["sequence"].apply(clean_sequence)
val_df["sequence_clean"]   = val_df["sequence"].apply(clean_sequence)
test_df["sequence_clean"]  = test_df["sequence"].apply(clean_sequence)

train_df[["sequence", "sequence_clean"]].head(2)
```

## Asking ESM-2 to read the proteins

Now comes the heart of the experiment.

Dr. Aris loads **ESM-2**, a transformer trained on protein sequences at scale.

Each protein will be passed through the model.  
For every amino acid position, ESM-2 will produce a contextual representation — a numerical description shaped not only by the residue itself, but by the sequence around it.

These token-level representations must then be compressed into a single fixed-length vector.

For this first baseline, Dr. Aris chooses **mean pooling**:
he averages the token embeddings across the valid amino acid positions.

This produces one embedding per protein — a compact signature of sequence meaning.

## Load tokenizer and ESM- 2 Model

```python
gc.collect()
if torch.cuda.is_available():
    torch.cuda.empty_cache()

print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

print("Loading ESM-2 model on CPU first...")
encoder_model = AutoModel.from_pretrained(
    MODEL_NAME,
    low_cpu_mem_usage=True
)

encoder_model.eval()

print("Moving model to device...")
encoder_model = encoder_model.to(DEVICE)

print("Loaded successfully.")
```

## Embedding helpers

```python
def mean_pool(last_hidden_state, attention_mask):
    mask = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
    masked_embeddings = last_hidden_state * mask
    summed = masked_embeddings.sum(dim=1)
    counts = mask.sum(dim=1).clamp(min=1e-9)
    return summed / counts

def embed_sequences(sequences, model, tokenizer, device="cpu", batch_size=8, max_len=1024):
    all_embeddings = []

    for i in tqdm(range(0, len(sequences), batch_size)):
        batch = sequences[i:i+batch_size]

        encoded = tokenizer(
            batch,
            padding=True,
            truncation=True,
            max_length=max_len,
            return_tensors="pt"
        )

        encoded = {k: v.to(device) for k, v in encoded.items()}

        with torch.no_grad():
            outputs = model(**encoded)
            pooled = mean_pool(outputs.last_hidden_state, encoded["attention_mask"])

        all_embeddings.append(pooled.cpu().numpy())

    return np.vstack(all_embeddings)
```

## Converting proteins into embeddings

One by one, and then batch by batch, the proteins are read by ESM-2.

For each sequence, the model produces a high-dimensional representation.
Mean pooling then condenses that representation into a single embedding vector.

These embeddings are the foundation of the downstream task.

If localization truly leaves a trace in sequence, then proteins from similar cellular compartments may begin to occupy similar regions of representation space.

## Generate Embeddings

```python
start_time = time.time()

X_train = embed_sequences(
    train_df["sequence_clean"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

X_val = embed_sequences(
    val_df["sequence_clean"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

X_test = embed_sequences(
    test_df["sequence_clean"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

embedding_time = time.time() - start_time

print("Embedding shapes:")
print("X_train:", X_train.shape)
print("X_val:  ", X_val.shape)
print("X_test: ", X_test.shape)

print(f"\nTotal embedding time: {embedding_time/60:.2f} minutes")
```

## Prepare Targets

```python
y_train = train_df["label"].values
y_val   = val_df["label"].values
y_test  = test_df["label"].values

print("y_train shape:", y_train.shape)
print("Number of classes:", len(label_encoder.classes_))
```

## Teaching a downstream classifier

ESM-2 has now transformed every protein into a numerical embedding.

But embeddings alone do not produce predictions.
To test whether localization information is present in them, Dr. Aris places a lightweight classifier on top.

He begins with a simple baseline:
**Logistic Regression**.

This is an important choice.

If a relatively simple classifier performs well, it suggests that the embeddings themselves already contain meaningful information about cellular destination.

```python
if CLASSIFIER_NAME == "logreg":
    clf = LogisticRegression(
        max_iter=3000,
        random_state=RANDOM_STATE,
        class_weight="balanced"
    )

elif CLASSIFIER_NAME == "mlp":
    clf = MLPClassifier(
        hidden_layer_sizes=(512, 256),
        activation="relu",
        solver="adam",
        alpha=1e-4,
        batch_size=64,
        learning_rate_init=1e-3,
        max_iter=100,
        random_state=RANDOM_STATE
    )

else:
    raise ValueError("Unsupported CLASSIFIER_NAME. Use 'logreg' or 'mlp'.")

print(clf)
```

## Train Classifier

```python
train_start = time.time()
clf.fit(X_train, y_train)
train_time = time.time() - train_start

print(f"Training time: {train_time:.2f} seconds")
```

## Validation Evaluation

```python
val_preds = clf.predict(X_val)

val_acc = accuracy_score(y_val, val_preds)
val_macro_f1 = f1_score(y_val, val_preds, average="macro")
val_weighted_f1 = f1_score(y_val, val_preds, average="weighted")

print("Validation Metrics")
print("------------------")
print(f"Accuracy     : {val_acc:.4f}")
print(f"Macro F1     : {val_macro_f1:.4f}")
print(f"Weighted F1  : {val_weighted_f1:.4f}")
```

## Test Evaluation

```python
test_preds = clf.predict(X_test)

test_acc = accuracy_score(y_test, test_preds)
test_macro_f1 = f1_score(y_test, test_preds, average="macro")
test_weighted_f1 = f1_score(y_test, test_preds, average="weighted")

print("Test Metrics")
print("------------")
print(f"Accuracy     : {test_acc:.4f}")
print(f"Macro F1     : {test_macro_f1:.4f}")
print(f"Weighted F1  : {test_weighted_f1:.4f}")
```

## Looking beyond accuracy

Dr. Aris never trusts a single number.

Accuracy is useful, but it can hide class imbalance.
So he also examines:

- **Macro F1**, which treats each localization class equally
- **Weighted F1**, which reflects the class distribution
- and the **confusion matrix**, which reveals which compartments are easily recognized and which tend to be confused

This way, the notebook does not merely ask whether the model works,
but how it behaves biologically.

## Classification Report

```python
print(classification_report(
    y_test,
    test_preds,
    target_names=label_encoder.classes_
))
```

## Confusion Matrix

```python
cm = confusion_matrix(y_test, test_preds)

fig, ax = plt.subplots(figsize=(8, 8))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=label_encoder.classes_
)
disp.plot(ax=ax, xticks_rotation=45, colorbar=False)
plt.title("Confusion Matrix - ESM-2 Subcellular Localization")
plt.tight_layout()
plt.show()
```

## Normalized confusion Matrix

```python
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(cm_norm, aspect="auto")

ax.set_xticks(np.arange(len(label_encoder.classes_)))
ax.set_yticks(np.arange(len(label_encoder.classes_)))
ax.set_xticklabels(label_encoder.classes_, rotation=45, ha="right")
ax.set_yticklabels(label_encoder.classes_)

for i in range(cm_norm.shape[0]):
    for j in range(cm_norm.shape[1]):
        ax.text(j, i, f"{cm_norm[i, j]:.2f}", ha="center", va="center")

ax.set_xlabel("Predicted label")
ax.set_ylabel("True label")
ax.set_title("Normalized Confusion Matrix - ESM-2")
plt.tight_layout()
plt.show()
```

## PCA visualization

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=RANDOM_STATE)
X_test_2d = pca.fit_transform(X_test)

pca_df = pd.DataFrame({
    "PC1": X_test_2d[:, 0],
    "PC2": X_test_2d[:, 1],
    "label": [id_to_label[i] for i in y_test]
})

plt.figure(figsize=(8, 6))
for label in sorted(pca_df["label"].unique()):
    subset = pca_df[pca_df["label"] == label]
    plt.scatter(subset["PC1"], subset["PC2"], s=15, alpha=0.7, label=label)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of ESM-2 Test Embeddings by Subcellular Localization")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

## Interpreting the result biologically

Dr. Aris now steps away from the code and returns to biology.

If ESM-2 embeddings support strong localization prediction, then the model has likely captured sequence features associated with cellular targeting and compartment-specific protein architecture.

This is biologically sensible:

- membrane proteins often contain hydrophobic transmembrane regions,
- extracellular proteins may contain secretion-associated signals,
- mitochondrial proteins may carry transport-related sequence features,
- nuclear proteins may bear localization motifs,
- and cytoplasmic proteins may lack many of these specialized signals.

The confusion matrix is especially revealing.

If two classes are repeatedly confused, this may mean:
- their sequence-level signals overlap,
- their targeting cues are subtle,
- or the embedding space does not cleanly separate them.

In every case, the model’s successes and failures offer insight.

## Imports

```python
import os
import re
import gc
import time
import random
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)

import torch
from transformers import T5Tokenizer, T5EncoderModel
from tqdm.auto import tqdm
```

```python
RANDOM_STATE = 42

DATA_DIR = "./subcellular_localization_data_prep"
TRAIN_PATH = f"{DATA_DIR}/train.csv"
VAL_PATH   = f"{DATA_DIR}/val.csv"
TEST_PATH  = f"{DATA_DIR}/test.csv"

MODEL_NAME = "Rostlab/prot_t5_xl_half_uniref50-enc"
MAX_LEN = 1024

# ProtT5 is memory-heavy
BATCH_SIZE = 1

CLASSIFIER_NAME = "logreg"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

print("Device:", DEVICE)
print("Model:", MODEL_NAME)
```

## Load Dataset

```python
train_df = pd.read_csv(TRAIN_PATH)
val_df   = pd.read_csv(VAL_PATH)
test_df  = pd.read_csv(TEST_PATH)

print(train_df.shape, val_df.shape, test_df.shape)
```

## Label Encoding

```python
label_encoder = LabelEncoder()
label_encoder.fit(train_df["localization"])

train_df["label"] = label_encoder.transform(train_df["localization"])
val_df["label"]   = label_encoder.transform(val_df["localization"])
test_df["label"]  = label_encoder.transform(test_df["localization"])

id_to_label = {i: l for i, l in enumerate(label_encoder.classes_)}
```

## Sequence formatting

```python
def clean_sequence(seq):
    seq = str(seq).upper().strip()
    seq = re.sub(r"[^ACDEFGHIKLMNPQRSTVWYUZOBX]", "", seq)
    seq = re.sub(r"[UZOB]", "X", seq)
    return seq

def format_for_prott5(seq):
    return " ".join(list(seq))

train_df["sequence_fmt"] = train_df["sequence"].apply(clean_sequence).apply(format_for_prott5)
val_df["sequence_fmt"]   = val_df["sequence"].apply(clean_sequence).apply(format_for_prott5)
test_df["sequence_fmt"]  = test_df["sequence"].apply(clean_sequence).apply(format_for_prott5)
```

## Loading ProtT5 carefully

ProtT5 is a large model.

If loaded naively, it may cause the kernel to crash due to memory spikes.
So Dr. Aris takes a careful approach:

- load model on CPU first
- reduce peak memory usage
- then move to GPU

```python
gc.collect()
if torch.cuda.is_available():
    torch.cuda.empty_cache()

print("Loading tokenizer...")
tokenizer = T5Tokenizer.from_pretrained(
    MODEL_NAME,
    do_lower_case=False
)

print("Loading ProtT5 encoder...")
encoder_model = T5EncoderModel.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    low_cpu_mem_usage=True,
    use_safetensors=True
)

encoder_model.eval()
encoder_model = encoder_model.to(DEVICE)

print("Loaded successfully.")
print("Model dtype:", next(encoder_model.parameters()).dtype)
```

## Converting proteins into ProtT5 embeddings

Now that ProtT5 is loaded, Dr. Aris asks it to read each protein sequence.

For every amino acid position, the encoder produces a contextual representation.  
These token-level representations are then combined into a single fixed-length vector using mean pooling.

This vector becomes the protein’s embedding — a compact summary of the biological information encoded in sequence.

```python
def mean_pool(last_hidden_state, attention_mask):
    mask = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
    masked_embeddings = last_hidden_state * mask
    summed = masked_embeddings.sum(dim=1)
    counts = mask.sum(dim=1).clamp(min=1e-9)
    return summed / counts


def embed_sequences(sequences, model, tokenizer, device="cpu", batch_size=1, max_len=1024):
    all_embeddings = []

    for i in tqdm(range(0, len(sequences), batch_size)):
        batch = sequences[i:i + batch_size]

        encoded = tokenizer(
            batch,
            padding=True,
            truncation=True,
            max_length=max_len,
            return_tensors="pt"
        )

        encoded = {k: v.to(device) for k, v in encoded.items()}

        with torch.no_grad():
            outputs = model(**encoded)
            pooled = mean_pool(outputs.last_hidden_state, encoded["attention_mask"])

        all_embeddings.append(pooled.cpu().numpy())

        # small cleanup
        del encoded, outputs, pooled
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    return np.vstack(all_embeddings)
```

## Creating embeddings for the training, validation, and test sets

Dr. Aris now passes the proteins through ProtT5 split by split.

Because ProtT5 is memory-intensive, he proceeds carefully, using a small batch size and clearing memory along the way.

By the end of this step, each protein will be represented by a numerical embedding ready for downstream classification.

## Generate Embeddings

```python
start_time = time.time()

X_train = embed_sequences(
    train_df["sequence_fmt"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

X_val = embed_sequences(
    val_df["sequence_fmt"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

X_test = embed_sequences(
    test_df["sequence_fmt"].tolist(),
    model=encoder_model,
    tokenizer=tokenizer,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    max_len=MAX_LEN
)

embedding_time = time.time() - start_time

print("Embedding shapes:")
print("X_train:", X_train.shape)
print("X_val:  ", X_val.shape)
print("X_test: ", X_test.shape)
print(f"\nTotal embedding time: {embedding_time/60:.2f} minutes")
```

## Prepare Targets

```python
y_train = train_df["label"].values
y_val   = val_df["label"].values
y_test  = test_df["label"].values

print("y_train shape:", y_train.shape)
print("Number of classes:", len(label_encoder.classes_))
```

## Teaching a downstream classifier

ProtT5 has now transformed each protein into an embedding.

But embeddings alone do not make predictions.  
To test whether localization information is encoded in them, Dr. Aris now places a lightweight classifier on top.

He begins with Logistic Regression — a simple but powerful baseline.

If this classifier performs well, it suggests that ProtT5 embeddings already contain meaningful signals about cellular destination.

## Train Classifier

```python
clf = LogisticRegression(
    max_iter=3000,
    random_state=RANDOM_STATE,
    class_weight="balanced"
)

train_start = time.time()
clf.fit(X_train, y_train)
train_time = time.time() - train_start

print(f"Training time: {train_time:.2f} seconds")
```

## Validation Evalution

```python
val_preds = clf.predict(X_val)

val_acc = accuracy_score(y_val, val_preds)
val_macro_f1 = f1_score(y_val, val_preds, average="macro")
val_weighted_f1 = f1_score(y_val, val_preds, average="weighted")

print("Validation Metrics")
print("------------------")
print(f"Accuracy     : {val_acc:.4f}")
print(f"Macro F1     : {val_macro_f1:.4f}")
print(f"Weighted F1  : {val_weighted_f1:.4f}")
```

## Test Evaluation

```python
test_preds = clf.predict(X_test)

test_acc = accuracy_score(y_test, test_preds)
test_macro_f1 = f1_score(y_test, test_preds, average="macro")
test_weighted_f1 = f1_score(y_test, test_preds, average="weighted")

print("Test Metrics")
print("------------")
print(f"Accuracy     : {test_acc:.4f}")
print(f"Macro F1     : {test_macro_f1:.4f}")
print(f"Weighted F1  : {test_weighted_f1:.4f}")
```

## Looking beyond a single score

Dr. Aris does not rely on accuracy alone.

Some localization classes may be easier to separate than others, and class imbalance can hide weaknesses behind a single number.

So he also examines:
- Macro F1, which treats each class equally
- Weighted F1, which accounts for class frequency
- and the confusion matrix, which reveals where the model tends to confuse one compartment with another

```python
print(classification_report(
    y_test,
    test_preds,
    target_names=label_encoder.classes_
))
```

## Confusion Matrix

```python
cm = confusion_matrix(y_test, test_preds)

fig, ax = plt.subplots(figsize=(8, 8))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=label_encoder.classes_
)
disp.plot(ax=ax, xticks_rotation=45, colorbar=False)
plt.title("Confusion Matrix - ProtT5 Subcellular Localization")
plt.tight_layout()
plt.show()
```

## Normalized Confusion Matrix

```python
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(cm_norm, aspect="auto")

ax.set_xticks(np.arange(len(label_encoder.classes_)))
ax.set_yticks(np.arange(len(label_encoder.classes_)))
ax.set_xticklabels(label_encoder.classes_, rotation=45, ha="right")
ax.set_yticklabels(label_encoder.classes_)

for i in range(cm_norm.shape[0]):
    for j in range(cm_norm.shape[1]):
        ax.text(j, i, f"{cm_norm[i, j]:.2f}", ha="center", va="center")

ax.set_xlabel("Predicted label")
ax.set_ylabel("True label")
ax.set_title("Normalized Confusion Matrix - ProtT5")
plt.tight_layout()
plt.show()
```

## PCA Plot

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=RANDOM_STATE)
X_test_2d = pca.fit_transform(X_test)

pca_df = pd.DataFrame({
    "PC1": X_test_2d[:, 0],
    "PC2": X_test_2d[:, 1],
    "label": [id_to_label[i] for i in y_test]
})

plt.figure(figsize=(8, 6))
for label in sorted(pca_df["label"].unique()):
    subset = pca_df[pca_df["label"] == label]
    plt.scatter(subset["PC1"], subset["PC2"], s=15, alpha=0.7, label=label)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of ProtT5 Test Embeddings by Subcellular Localization")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

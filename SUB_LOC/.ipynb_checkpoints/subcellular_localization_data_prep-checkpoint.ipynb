Some proteins remain in the cytoplasm, carrying out the busy chemistry of the cell’s interior.  
Some travel to the nucleus, where information is stored and regulated.  
Some settle into membranes, becoming gates, sensors, or anchors.  
Others are secreted outward, acting beyond the boundaries of the cell.

So in this notebook, Dr. Aris begins by building the dataset itself.

Before any model can learn localization, the data must be gathered carefully:
- from trusted protein records,
- with clean sequences,
- with interpretable localization labels,
- and with enough examples per class to support a fair learning task.

This notebook is devoted entirely to that preparation.

## Goal of this notebook

In this notebook, Dr. Aris will:

1. Retrieve protein entries from UniProt
2. Keep reviewed, non-fragment proteins
3. Extract sequence and subcellular location annotations
4. Convert raw location annotations into clean labels
5. Keep only single-label proteins for the first experiment
6. Filter rare classes
7. Produce train, validation, and test CSV files

The final output will be a clean dataset ready for downstream transformer-based modeling.

```python
import re
import time
import math
import json
import random
import requests
import warnings
import os
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from io import StringIO
from sklearn.model_selection import train_test_split
```

```python
# =========================
# Configuration
# =========================
RANDOM_STATE = 42
MAX_RECORDS = 20000           # set None if you want to keep all retrieved rows
MIN_SEQ_LEN = 30
MAX_SEQ_LEN = 2000
MIN_SAMPLES_PER_CLASS = 200
TOP_K_CLASSES = 5

OUT_DIR = "./subcellular_localization_data_prep"

random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)

print("Configuration loaded.")
```

## Why start from UniProt?

Dr. Aris wanted a dataset that was not merely large, but biologically trustworthy.

So he turned to UniProt and asked for proteins that were:
- **reviewed**, so their annotations would be curated,
- **non-fragment**, so their sequences would more likely represent full proteins,
- and associated with **subcellular location annotations**, because cellular destination was the very signal he hoped to study.

For this first version, he chose quality and interpretability over sheer scale.

```python
# UniProt REST API endpoint
BASE_URL = "https://rest.uniprot.org/uniprotkb/search"

# Query:
# - reviewed proteins
# - not fragments
# - with subcellular location comment
QUERY = "(reviewed:true) AND (fragment:false) AND (cc_scl_term:*)"

# Return fields:
# accession, protein_name, organism_name, length, sequence, cc_subcellular_location
FIELDS = [
    "accession",
    "protein_name",
    "organism_name",
    "length",
    "sequence",
    "cc_subcellular_location"
]

params = {
    "query": QUERY,
    "format": "tsv",
    "fields": ",".join(FIELDS),
    "size": 500
}

print("Query:", QUERY)
print("Fields:", FIELDS)
```

## Retrieving protein records

Dr. Aris now sends his request to UniProt.

He asks not for every possible detail, but only for the pieces needed to build the localization dataset:
- accession
- protein name
- organism
- sequence length
- sequence
- subcellular location annotation

These will serve as the raw ingredients of the task.

```python
def fetch_uniprot_tsv(query, fields, batch_size=500, max_records=None):
    """
    Fetch TSV results from UniProt REST API with pagination.
    """
    session = requests.Session()
    url = "https://rest.uniprot.org/uniprotkb/search"
    
    params = {
        "query": query,
        "format": "tsv",
        "fields": ",".join(fields),
        "size": batch_size
    }

    all_chunks = []
    fetched_rows = 0
    next_url = url

    while next_url:
        response = session.get(next_url, params=params if next_url == url else None, timeout=120)
        response.raise_for_status()

        text = response.text.strip()
        if text:
            chunk_df = pd.read_csv(StringIO(text), sep="\t")
            all_chunks.append(chunk_df)
            fetched_rows += len(chunk_df)

        if max_records is not None and fetched_rows >= max_records:
            break

        link_header = response.headers.get("Link")
        next_url = None

        if link_header and 'rel="next"' in link_header:
            match = re.search(r'<(.+?)>; rel="next"', link_header)
            if match:
                next_url = match.group(1)

        params = None  # only needed for the first request

    if not all_chunks:
        return pd.DataFrame()

    df = pd.concat(all_chunks, ignore_index=True)

    if max_records is not None:
        df = df.iloc[:max_records].copy()

    return df
```

```python
start = time.time()

raw_df = fetch_uniprot_tsv(
    query=QUERY,
    fields=FIELDS,
    batch_size=500,
    max_records=MAX_RECORDS
)

end = time.time()

print(f"Fetched shape: {raw_df.shape}")
print(f"Time taken: {(end - start)/60:.2f} minutes")
raw_df.head()
```

```python
os.makedirs(OUT_DIR, exist_ok=True)

raw_path = f"{OUT_DIR}/uniprot_raw_subcellular.tsv"
raw_df.to_csv(raw_path, sep="\t", index=False)

print("Saved raw file to:", raw_path)
```

```python
print(raw_df.columns.tolist())
```

```python
df = raw_df.copy()

rename_map = {
    "Entry": "accession",
    "Protein names": "protein_name",
    "Organism": "organism_name",
    "Length": "length",
    "Sequence": "sequence",
    "Subcellular location [CC]": "subcellular_location"
}

df = df.rename(columns=rename_map)

expected_cols = [
    "accession",
    "protein_name",
    "organism_name",
    "length",
    "sequence",
    "subcellular_location"
]

print("Available columns after rename:")
print(df.columns.tolist())
```

## Cleaning the sequences

Before the proteins can be organized into cellular destinations, Dr. Aris first checks whether the sequences themselves are usable.

He removes:
- missing entries,
- duplicate proteins,
- extremely short sequences,
- extremely long sequences,
- and any unexpected sequence characters.

This ensures that the future embedding model will see clean biological input rather than avoidable noise.

```python
df = df[expected_cols].copy()

# Drop missing core fields
df = df.dropna(subset=["sequence", "subcellular_location"]).reset_index(drop=True)

# Sequence cleaning
def clean_sequence(seq):
    seq = str(seq).upper().strip()
    seq = re.sub(r"[^ACDEFGHIKLMNPQRSTVWYUZOBX]", "", seq)
    seq = re.sub(r"[UZOB]", "X", seq)
    return seq

df["sequence"] = df["sequence"].apply(clean_sequence)
df["seq_len"] = df["sequence"].str.len()

# Length filtering
df = df[df["seq_len"] >= MIN_SEQ_LEN]
df = df[df["seq_len"] <= MAX_SEQ_LEN]

# Remove exact duplicate sequences
df = df.drop_duplicates(subset=["sequence"]).reset_index(drop=True)

print("Shape after sequence cleaning:", df.shape)
df.head()
```

```python
pd.set_option("display.max_colwidth", 200)
df[["accession", "subcellular_location"]].head(10)
```

## Turning annotation text into usable localization labels

The raw subcellular location field is rich, but not yet ready for classification.

A protein entry may mention:
- multiple possible locations,
- membrane topology,
- orientation details,
- or free-text notes about evidence and biological context.

Dr. Aris now performs a careful simplification.

He extracts the main location terms and maps them into a cleaner set of classes such as:
- Cytoplasm
- Nucleus
- Membrane
- Extracellular
- Mitochondrion

For the first experiment, he keeps only proteins that can be assigned one clear primary label.

```python
CANONICAL_LABELS = {
    "cytoplasm": "Cytoplasm",
    "cytosol": "Cytoplasm",
    "nucleus": "Nucleus",
    "nucleolus": "Nucleus",
    "nucleoplasm": "Nucleus",
    "membrane": "Membrane",
    "cell membrane": "Membrane",
    "plasma membrane": "Membrane",
    "extracellular": "Extracellular",
    "secreted": "Extracellular",
    "mitochondrion": "Mitochondrion",
    "mitochondria": "Mitochondrion",
    "endoplasmic reticulum": "Endoplasmic Reticulum",
    "er": "Endoplasmic Reticulum",
    "golgi": "Golgi",
    "lysosome": "Lysosome",
    "lysosomal": "Lysosome",
    "peroxisome": "Peroxisome",
    "chloroplast": "Chloroplast"
}

priority_order = [
    "Cytoplasm",
    "Nucleus",
    "Membrane",
    "Extracellular",
    "Mitochondrion",
    "Endoplasmic Reticulum",
    "Golgi",
    "Lysosome",
    "Peroxisome",
    "Chloroplast"
]

def extract_candidate_labels(text):
    """
    Extract canonical localization labels from UniProt subcellular text.
    """
    if pd.isna(text):
        return []

    txt = str(text).lower()

    found = set()
    for key, value in CANONICAL_LABELS.items():
        if key in txt:
            found.add(value)

    return sorted(found)

def choose_primary_label(labels):
    """
    Keep only one label according to priority order.
    """
    if not labels:
        return None
    for lbl in priority_order:
        if lbl in labels:
            return lbl
    return labels[0]
```

```python
df["candidate_labels"] = df["subcellular_location"].apply(extract_candidate_labels)
df["n_candidate_labels"] = df["candidate_labels"].apply(len)
df["primary_label"] = df["candidate_labels"].apply(choose_primary_label)

print(df[["subcellular_location", "candidate_labels", "primary_label"]].head(15))
```

```python
print("Rows with at least one mapped label:", (df["primary_label"].notna()).sum())
print("Rows with zero mapped labels:", (df["primary_label"].isna()).sum())
print("\nNumber of candidate labels per protein:")
print(df["n_candidate_labels"].value_counts().sort_index())
```

```python
single_label_df = df[df["n_candidate_labels"] == 1].copy().reset_index(drop=True)

print("Single-label dataset shape:", single_label_df.shape)
print(single_label_df["primary_label"].value_counts())
```

```python
plt.figure(figsize=(10, 5))
single_label_df["primary_label"].value_counts().plot(kind="bar")
plt.ylabel("Count")
plt.title("Subcellular Localization Class Distribution (Single-label Only)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

## Building the first benchmark set

A good first benchmark should be clear, balanced enough to learn from, and simple enough to interpret.

So Dr. Aris keeps only the most frequent classes and removes labels with too few proteins.  
This avoids building a dataset where the model sees thousands of examples of one compartment but only a handful from another.

For the first experiment, simplicity is not a limitation — it is a strength.

```python
class_counts = single_label_df["primary_label"].value_counts()

top_classes = class_counts.head(TOP_K_CLASSES).index.tolist()
filtered_df = single_label_df[single_label_df["primary_label"].isin(top_classes)].copy()

# Enforce minimum size
valid_classes = filtered_df["primary_label"].value_counts()
valid_classes = valid_classes[valid_classes >= MIN_SAMPLES_PER_CLASS].index.tolist()

filtered_df = filtered_df[filtered_df["primary_label"].isin(valid_classes)].copy().reset_index(drop=True)

print("Final class counts:")
print(filtered_df["primary_label"].value_counts())
print("Final dataset shape:", filtered_df.shape)
```

```python
BALANCE_CLASSES = False

if BALANCE_CLASSES:
    min_count = filtered_df["primary_label"].value_counts().min()
    filtered_df = (
        filtered_df.groupby("primary_label", group_keys=False)
        .apply(lambda x: x.sample(min_count, random_state=RANDOM_STATE))
        .reset_index(drop=True)
    )
    print("Balanced class counts:")
    print(filtered_df["primary_label"].value_counts())
```

```python
final_df = filtered_df[[
    "accession",
    "protein_name",
    "organism_name",
    "sequence",
    "seq_len",
    "primary_label"
]].copy()

final_df = final_df.rename(columns={"primary_label": "localization"})

print(final_df.shape)
final_df.head()
```

```python
plt.figure(figsize=(10, 5))
final_df["localization"].value_counts().plot(kind="bar")
plt.ylabel("Count")
plt.title("Final Class Distribution for Subcellular Localization Dataset")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
clean_path = f"{OUT_DIR}/subcellular_localization_clean_dataset.csv"
final_df.to_csv(clean_path, index=False)

print("Saved clean dataset to:", clean_path)
```

## Splitting the dataset

Now that the proteins are organized into a cleaner set of cellular destinations, Dr. Aris prepares them for learning.

He separates the data into:
- a training set, where the classifier will learn,
- a validation set, where decisions can be tuned,
- and a test set, where the final evaluation will be performed fairly.

To preserve the biological balance across classes, he uses a **stratified split**.

```python
train_df, temp_df = train_test_split(
    final_df,
    test_size=0.30,
    stratify=final_df["localization"],
    random_state=RANDOM_STATE
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["localization"],
    random_state=RANDOM_STATE
)

print("Train shape:", train_df.shape)
print("Val shape:  ", val_df.shape)
print("Test shape: ", test_df.shape)

print("\nTrain distribution:")
print(train_df["localization"].value_counts())
```

```python
train_path = f"{OUT_DIR}/train.csv"
val_path   = f"{OUT_DIR}/val.csv"
test_path  = f"{OUT_DIR}/test.csv"

train_df.to_csv(train_path, index=False)
val_df.to_csv(val_path, index=False)
test_df.to_csv(test_path, index=False)

print("Saved:")
print(train_path)
print(val_path)
print(test_path)
```

```python
summary_df = pd.DataFrame({
    "split": ["train", "val", "test"],
    "n_samples": [len(train_df), len(val_df), len(test_df)],
    "n_classes": [
        train_df["localization"].nunique(),
        val_df["localization"].nunique(),
        test_df["localization"].nunique()
    ],
    "avg_seq_len": [
        train_df["seq_len"].mean(),
        val_df["seq_len"].mean(),
        test_df["seq_len"].mean()
    ]
})

summary_df
```
